# 不同温度下知识蒸馏softmax及预测准确率每轮epoch可视化

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

%config InlineBackend.figure_format = 'retina'
```

```python
logits = np.array([-3,2,6,10])
```

```python
x=np.array(['car',  'cat','donkey', 'horse'])
# plt.plot(logits,label='logits')
plt.bar(x,logits,color='grey',width=0.6)
plt.legend()
# plt.show()
###增加散点样式
plt.savefig('./horsetargets.png')
```

```python
softmax_1 = np.exp(logits) / sum(np.exp(logits))
```

```python
softmax_1
```

```python
x=np.array(['car',  'cat','donkey', 'horse'])
plt.plot(x,softmax_1,'b*-',label='softmax T=1')
plt.legend()
# plt.show()
plt.savefig('./softmax1.png')
```

```python
x=np.array(['car',  'cat','donkey', 'horse'])
plt.plot(x,softmax_1,'b*-',label='T=1')

T = 3
softmax_3 = np.exp(logits/T) / sum(np.exp(logits/T))
plt.plot(x,softmax_3,'r^-',label='T=3')

T = 5
softmax_5 = np.exp(logits/T) / sum(np.exp(logits/T))
plt.plot(x,softmax_5,'gx-',label='T=5')

T = 10
softmax_10 = np.exp(logits/T) / sum(np.exp(logits/T))
plt.plot(x,softmax_10,'md-',label='T=10')

T = 100
softmax_100 = np.exp(logits/T) / sum(np.exp(logits/T))
plt.plot(x,softmax_100,'c.-',label='T=100')

plt.xticks(np.arange(4), ['Cat', 'Dog', 'Donkey', 'Horse']) 
plt.legend()
# plt.show()
plt.savefig('./softmax100.png')
```
